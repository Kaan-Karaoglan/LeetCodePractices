```python
class Solution:
    def reverse(self, x: int) -> int:
        reversed_s = ""
        str_x = str(x)
        reversed_s = str_x[::-1]
        if str_x.__contains__("-") == True:
            reversed_s = "-" + reversed_s[:-1]
            if int(reversed_s) < -2147483648:
                reversed_s = "0"
        elif len(reversed_s) == 1:
            reversed_s = str_x
        elif int(reversed_s) > 2147483648:
            reversed_s = "0"
        return int(reversed_s)
```
